### **Importing the required libraries and Iris data set**



```
import pandas as pd
import sklearn.datasets as datasets
iris=datasets.load_iris()
```

### **Loading the iris dataset and its attributes to a data frame**

```
ir=pd.DataFrame(iris.data,columns=iris.feature_names)
print("\033[1mHead of Iris data set\033[0m")
print(ir.head(10))
print("\033[1mTail of iris data set\033[0m")
print(ir.tail(10))
```

### **Knowing the Data Set**

```
ir.info()
```

```
ir.describe()
```

#### **Checking for any null entries**

```
ir.isnull().sum()
```

### **Loading the Target attribute of Iris data set**

```
Y = iris.target
Y
```

```
from sklearn.model_selection import train_test_split
ir_train, ir_test, y_train, y_test = train_test_split(ir, Y, test_size=0.33, random_state=0)
```

### **Building the decision tree using train data**

```
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(ir_train,y_train)
```

### **Using the decision tree to predict target attributes of test data set of Iris data set**

```
y_predict = dtc.predict(ir_test)
y_predict
```

### **Comparing the predicted and actual values**

```
com = pd.DataFrame({'Actual': y_test,'Predicted':y_predict})
com.head(10)
```

**Confusion Matrix is a performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.**

4.   False Negative: (Type 2 Error) You predicted negative and it’s false.



### **Creating the confusion matrix**

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

#### **From the above confusion matrix,we can conclude that the there are very few wrong prediction and the decision tree is reliable for predicting thr right class accordingly**

### **Ploting the decision tree graphically**

```
from sklearn import tree
import matplotlib.pyplot as plt
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtc, feature_names = fn, class_names = cn, filled = True);
```
